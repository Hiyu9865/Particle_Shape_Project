```python
import trimesh
import numpy as np
import os
import random
import re
from typing import Tuple, Optional

# --- 設定 ---
PROJECT_ROOT = '/content/drive/MyDrive/SHAPE_AI_Project'
OUTPUT_DIR = os.path.join(PROJECT_ROOT, '01_3D_Models')
os.makedirs(OUTPUT_DIR, exist_ok=True)
# ----------

def generate_constrained_base_mesh(s: float, i_axis: float, l: float) -> trimesh.Trimesh:
    """
    ターゲット軸長 (S, I, L) に制約されつつ、球体ではない不規則な基底メッシュを生成する。
    """
    # 1. 高解像度の球のメッシュを作成
    mesh = trimesh.creation.icosphere(subdivisions=4)

    # 2. 強いランダムノイズを加え、球体形状を破綻させる
    # ノイズの強さは軸長の平均に基づいてランダムに決定
    avg_axis_length = (s + i_axis + l) / 3
    noise_magnitude = random.uniform(avg_axis_length * 0.05, avg_axis_length * 0.15)

    # 頂点法線を取得し、ランダムなノイズを法線方向に適用
    vertex_normals = mesh.vertex_normals
    displacement = np.random.uniform(-noise_magnitude, noise_magnitude, len(mesh.vertices))
    mesh.vertices += (vertex_normals.T * displacement).T

    # 3. 指定された軸長でスケーリングして制約を与える
    radii = [s / 2, i_axis / 2, l / 2]
    transform = np.diag([radii[0], radii[1], radii[2], 1.0])
    mesh = mesh.apply_transform(transform)

    mesh.process()
    return mesh


def generate_random_convex_hull_particle(target_L: float, file_id: int):
    """
    楕円体に基づかず、ランダムな点の凸包から真にランダムな形状を生成する関数。
    """
    num_points = random.randint(30, 60)

    # -L/2 から L/2 の範囲でランダムな点を生成
    points = np.random.uniform(low=-target_L / 2, high=target_L / 2, size=(num_points, 3))

    # ⚠️ 修正点: trimesh.points.hull がない環境に対応するため、trimesh.points.convex_hull を使用
    try:
        mesh = trimesh.convex.convex_hull(points)
    except AttributeError:
        # 古いバージョンや環境に対応するため、代替ロジックを試す
        from scipy.spatial import ConvexHull
        hull = ConvexHull(points)
        mesh = trimesh.Trimesh(vertices=points[hull.vertices], faces=hull.simplices)
        mesh.process()


    # ノイズを追加して表面の不規則性を強調
    mesh.vertices += np.random.normal(0, target_L * 0.01, mesh.vertices.shape)

    # ファイル名のために、生成されたメッシュの最小バウンディングボックスの軸長を再計算
    S, I, L_calc = np.sort(mesh.bounding_box.extents)

    filename = f"particle_{file_id}_S{S:.1f}I{I:.1f}L{L_calc:.1f}_Hull.stl"
    output_path = os.path.join(OUTPUT_DIR, filename)
    with open(output_path, 'wb') as f:
        mesh.export(f, file_type='stl')

    print(f"✅ 生成完了 (ランダム凸包): {filename}")
    return S, I, L_calc

def generate_base_ellipsoid_smooth(s, i_axis, l, subdivisions=4):
    """基本となる滑らかな楕円体のメッシュを生成する（ノイズなし）。"""
    radii = [s / 2, i_axis / 2, l / 2]
    sphere = trimesh.creation.icosphere(subdivisions=subdivisions)
    transform = np.diag([radii[0], radii[1], radii[2], 1.0])
    return sphere.apply_transform(transform)

def generate_satellite_particle(s, i_axis, l, file_id, satellite_count: int, ratio: float = 0.2):
    """
    衛星状（クラスター状）の粒子を生成する関数。コアは滑らかな楕円体を使用する。
    """
    # 1. コア粒子を生成 (滑らかな楕円体/球体を使用)
    core_mesh = generate_base_ellipsoid_smooth(s, i_axis, l)
    core_mesh.process() # 法線とジオメトリを計算

    all_meshes = [core_mesh]
    avg_axis_length = (s + i_axis + l) / 3
    sat_radius = avg_axis_length * ratio / 2

    for _ in range(satellite_count):
        # 衛星を生成 (小さな球体)
        sat_mesh = trimesh.creation.icosphere(radius=sat_radius, subdivisions=2)

        # コアの表面のランダムな位置に衛星を配置
        surface_points = core_mesh.sample(1)
        if surface_points.shape[0] == 0: continue

        # 表面に最も近い面インデックスを取得
        nearest_result = core_mesh.nearest.on_surface(surface_points)
        closest_face_index = int(nearest_result[1][0])

        # 法線方向に沿って移動させて、くっつくように配置
        normal = core_mesh.face_normals[closest_face_index]
        position = surface_points[0] + normal * sat_radius * 0.8 # 80%食い込ませる

        sat_mesh.apply_translation(position)
        all_meshes.append(sat_mesh)

    # すべてのメッシュを結合
    cluster_mesh = trimesh.util.concatenate(all_meshes)
    cluster_mesh.process()

    # ファイル名のために、生成されたメッシュの軸長を計算
    S_calc, I_calc, L_calc = np.sort(cluster_mesh.bounding_box.extents)

    filename = f"particle_{file_id}_S{S_calc:.1f}I{I_calc:.1f}L{L_calc:.1f}_Sat.stl"
    output_path = os.path.join(OUTPUT_DIR, filename)
    with open(output_path, 'wb') as f:
        cluster_mesh.export(f, file_type='stl')

    print(f"✅ 生成完了 (衛星状クラスター: 球体コア): {filename}")
    return S_calc, I_calc, L_calc

def extract_sil_from_filename(filename):
    """
    ファイル名から短軸S, 中軸I, 長軸Lの値を抽出するヘルパー関数
    """
    match = re.search(r'S([\d.]+)\s*I([\d.]+)\s*L([\d.]+)', filename)
    if match:
        S = float(match.group(1))
        I = float(match.group(2))
        L = float(match.group(3))
        return S, I, L
    raise ValueError(f"Filename does not contain S, I, L parameters: {filename}")
```

```python
# --- モデル生成の実行 (不規則ベース, ランダム凸包, 衛星状を組み合わせてカバー) ---
print(f"3Dモデルを生成し、ディレクトリ {OUTPUT_DIR} に保存します...")

# 生成するランダム粒子の数 (検証動画の安定性のため、200個を推奨)
NUM_RANDOM_PARTICLES = 200

# 粒子のスケール範囲
MAX_SCALE = 15.0
MIN_SCALE = 5.0

particle_type_counts = {'Irregular_Base': 0, 'Random_Hull': 0, 'Satellite': 0}

for i in range(NUM_RANDOM_PARTICLES):
    # 1. 軸長のランダム生成 (全てのタイプで使用)
    L = random.uniform(MIN_SCALE, MAX_SCALE)
    I = random.uniform(L * 0.1, L)
    S = random.uniform(I * 0.1, I)

    # 2. 生成タイプをランダムに選択
    generation_type = random.choice(['Irregular_Base', 'Random_Hull', 'Satellite'])

    try:
        if generation_type == 'Irregular_Base':
            # 非楕円・不規則ベース + 軸制約粒子
            mesh = generate_constrained_base_mesh(S, I, L)

            S_calc, I_calc, L_calc = S, I, L
            filename_suffix = "_Irreg.stl"

            # STLエクスポート
            output_path = os.path.join(OUTPUT_DIR, f"particle_{i+1}_S{S_calc:.1f}I{I_calc:.1f}L{L_calc:.1f}{filename_suffix}")
            with open(output_path, 'wb') as f:
                mesh.export(f, file_type='stl')
            print(f"✅ 生成完了 (不規則ベース): particle_{i+1}_S{S_calc:.1f}...")

            particle_type_counts['Irregular_Base'] += 1

        elif generation_type == 'Random_Hull':
            # 軸制約に依存しない、真にランダムな凸包形状
            S_calc, I_calc, L_calc = generate_random_convex_hull_particle(L, i + 1)
            particle_type_counts['Random_Hull'] += 1

        elif generation_type == 'Satellite':
            # 衛星状粒子 (コアも不規則な基底を使用)
            satellite_count = random.randint(2, 5) # 2〜5個の衛星
            ratio = random.uniform(0.1, 0.4)      # 衛星のサイズ比率
            S_calc, I_calc, L_calc = generate_satellite_particle(S, I, L, i + 1, satellite_count, ratio)
            particle_type_counts['Satellite'] += 1

    except Exception as e:
        print(f"❌ モデル生成失敗 (ID {i+1}, Type {generation_type}): {e}")


print(f"\n--- {NUM_RANDOM_PARTICLES} 個の複合モデルの生成完了 ---")
print(f"内訳: {particle_type_counts}")
```

```python
import os
import pandas as pd
import trimesh
import sys
import numpy as np
import re # extract_sil_from_filename 関数を使用するため

# --- 1. 設定とモジュールパスの準備 ---
PROJECT_ROOT = "/content/drive/MyDrive/SHAPE_AI_Project"
sys.path.append(os.path.join(PROJECT_ROOT, "src"))

# 実行するモジュールをインポート (pyshapeの全関数)
try:
    from pyshape import convexity, sphericity_wadell, form_functions_2, surface_orientation_tensor
    # form_functions_1は不要 (convexityとsphericity_wadellを個別に使用)
    print("✅ 'pyshape' モジュールの全計算関数を正常にインポートしました。")
except ImportError as e:
    print(f"❌ インポートエラー: 'pyshape' モジュールが見つかりません。ZIPファイルの配置を確認してください。: {e}")
    raise

# STLファイルがあるフォルダ（入力）
INPUT_STL_DIR = os.path.join(PROJECT_ROOT, "01_3D_Models")
# ラベルデータ（CSV）の出力先
OUTPUT_CSV_PATH = os.path.join(PROJECT_ROOT, "02_GroundTruth_Labels", "comprehensive_particle_labels.csv")
# ----------------------------------------------------------------------


# 3. 計算の実行
results = []
file_list = sorted([f for f in os.listdir(INPUT_STL_DIR) if f.endswith('.stl')])

if not file_list:
    print("⚠️ エラー: STLファイルが見つかりませんでした。タスク1.2の実行を確認してください。")
else:
    print(f"\n--- {len(file_list)} 個のSTLファイルに対して包括的な形状計算を開始します ---")

    for filename in file_list:
        file_path = os.path.join(INPUT_STL_DIR, filename)

        try:
            # 1. 軸長パラメータの抽出
            # S, I, L は軸長ベースの特性計算に必須
            S, I, L = extract_sil_from_filename(filename)

            # 2. STLファイルを読み込み (trimeshを使用)
            mesh = trimesh.load_mesh(file_path)

            # 3. 必須パラメータの計算 (体積、面積、凸包体積)
            volume = mesh.volume
            area = mesh.area
            volume_convex_hull = mesh.convex_hull.volume

            # 4. pyshapeモジュールを使った形状特性の計算 (9種類)

            # 4.1. 体積・面積ベース
            wadell = sphericity_wadell(volume, area)
            conv = convexity(volume, volume_convex_hull)

            # 4.2. 軸長ベース (Krumbein Sphericity, Zingg, Potticary, Kong&Fonseca)
            spK, flP, elP, flK, elK, SIZ, ILZ = form_functions_2(S, I, L)

            # 4.3. 表面配向テンソルベース (Compactness, Flakiness, Rodness)
            C_ten, F_ten, R_ten, _, _ = surface_orientation_tensor(mesh.vertices, mesh.faces)


            # 5. 結果の記録 (全ての特性を追加)
            results.append({
                'File_Name': filename,
                'Volume_um3': volume,
                'Area_um2': area,
                'Convex_Hull_Volume_um3': volume_convex_hull,
                'Wadell_Sphericity': wadell,
                'Convexity': conv,
                'Krumbein_Sphericity': spK,
                'Zingg_S_I_Ratio': SIZ,
                'Zingg_I_L_Ratio': ILZ,
                'Potticary_Flatness': flP,
                'Potticary_Elongation': elP,
                'KongFonseca_Flatness': flK,
                'KongFonseca_Elongation': elK,
                'Tensor_Compactness': C_ten,
                'Tensor_Flakiness': F_ten,
                'Tensor_Rodness': R_ten,
            })
            print(f"  [完了] {filename}: Wadell={wadell:.3f}, Conv={conv:.3f}, Zingg S/I={SIZ:.3f}")

        except Exception as e:
            # STLファイルが複雑すぎたり、凸包計算でエラーが出たりした場合
            print(f"  [失敗] {filename}: 計算中に予期せぬエラーが発生しました: {e}")

    # 6. 結果をCSVとして出力
    results_df = pd.DataFrame(results)
    results_df.to_csv(OUTPUT_CSV_PATH, index=False)

    print(f"\n✅ 全計算が完了しました。CSVファイルは {OUTPUT_CSV_PATH} に保存されました。")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# --- 設定 ---
PROJECT_ROOT = "/content/drive/MyDrive/SHAPE_AI_Project"
# タスク1.3で出力された包括的なラベルCSVファイル
INPUT_CSV_PATH = os.path.join(PROJECT_ROOT, '02_GroundTruth_Labels', 'comprehensive_particle_labels.csv')
# 出力画像ファイル名
OUTPUT_PLOT_PATH = os.path.join(PROJECT_ROOT, 'zingg_diagram_plot.png')
# --- 設定 ---

# 1. データの読み込み
try:
    df = pd.read_csv(INPUT_CSV_PATH)
    print(f"✅ ラベルデータ {INPUT_CSV_PATH} を読み込みました。")
except FileNotFoundError:
    print(f"❌ エラー: CSVファイルが見つかりません。タスク1.3を実行して {INPUT_CSV_PATH} が生成されているか確認してください。")
    raise

# 2. Zingg分類の境界値
ZINGG_BOUNDARY = 2/3 # 約 0.667

# 3. 形状の分類関数
def classify_zingg(row):
    # 横軸: S/I (Zingg_S_I_Ratio)
    # 縦軸: I/L (Zingg_I_L_Ratio)
    SI_ratio = row['Zingg_S_I_Ratio']
    IL_ratio = row['Zingg_I_L_Ratio']

    if SI_ratio > ZINGG_BOUNDARY and IL_ratio > ZINGG_BOUNDARY:
        return 'Compact (球状)', 'red'
    elif SI_ratio > ZINGG_BOUNDARY and IL_ratio <= ZINGG_BOUNDARY:
        return 'Prolate (棒状)', 'blue'
    elif SI_ratio <= ZINGG_BOUNDARY and IL_ratio > ZINGG_BOUNDARY:
        return 'Oblate (円盤状)', 'green'
    else:
        return 'Bladed (小判状)', 'gold'

# 4. 分類をデータフレームに追加
df[['Zingg_Class', 'Color']] = df.apply(classify_zingg, axis=1, result_type='expand')


# 5. プロットの実行
plt.figure(figsize=(8, 8))

# 散布図をプロット
for name, group in df.groupby('Zingg_Class'):
    plt.scatter(
        group['Zingg_S_I_Ratio'],
        group['Zingg_I_L_Ratio'],
        c=group['Color'].iloc[0], # グループの代表色を使用
        label=name,
        alpha=0.8,
        s=50 # 点のサイズ
    )

# 6. Zingg分類の境界線を追加
plt.axvline(x=ZINGG_BOUNDARY, color='k', linestyle='--', linewidth=1)
plt.axhline(y=ZINGG_BOUNDARY, color='k', linestyle='--', linewidth=1)

# 7. プロットの装飾
plt.title('Zingg Diagram Classification of Generated Particles', fontsize=14)
plt.xlabel('$S/I$ Ratio (Short Axis / Intermediate Axis)', fontsize=12)
plt.ylabel('$I/L$ Ratio (Intermediate Axis / Long Axis)', fontsize=12)
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend(title="Zingg Shape Class")
plt.gca().set_aspect('equal', adjustable='box') # アスペクト比を1:1に設定

# 8. 画像の保存
plt.savefig(OUTPUT_PLOT_PATH)
plt.close()

print(f"\n✅ Zinggダイアグラムのプロットが完了しました。画像は {OUTPUT_PLOT_PATH} に保存されました。")
print("生成された図を確認し、粒子が4つの象限に偏りなく分布していることを確認してください。")
```

```python
import os
import trimesh
import numpy as np
import cv2
import random
import pyrender
from tqdm import tqdm

# --- 設定 ---
PROJECT_ROOT = "/content/drive/MyDrive/SHAPE_AI_Project"
INPUT_STL_DIR = os.path.join(PROJECT_ROOT, '01_3D_Models')
OUTPUT_DIR = os.path.join(PROJECT_ROOT, '03_Simulation_Videos')
os.makedirs(OUTPUT_DIR, exist_ok=True)

# レンダリング要件
RESOLUTION = (100, 100) # 100x100px
FRAME_COUNT = 100
FPS = 30
NUM_PARTICLES = 1

# カメラとノイズ設定 (V31の設定を維持)
CAMERA_DISTANCE = 15.0
TRACKING_ERROR_SIGMA = 5 # ピクセル
LIGHT_INTENSITY = 10.0 # ⚠️ 修正点 1: 光の強度を上げて影のコントラストを強調
# --- 設定 ---

# 4. 動画書き出し設定 (OpenCV)
fourcc = cv2.VideoWriter_fourcc(*'mp4v')

file_list = [f for f in os.listdir(INPUT_STL_DIR) if f.endswith('.stl')]

print(f"\n--- Pyrender を使用してノイズ特化データを生成 (タスク2.1: V38 - バックライト最適化) ---")

NUM_VIDEOS_TO_GENERATE = 100
r = pyrender.OffscreenRenderer(RESOLUTION[0], RESOLUTION[1])

for video_index in range(NUM_VIDEOS_TO_GENERATE):
    stl_file = random.choice(file_list)
    input_path = os.path.join(INPUT_STL_DIR, stl_file)
    output_video_path = os.path.join(OUTPUT_DIR, f'noise_rot_video_{video_index:02d}_v38_backlit.mp4')

    # 粒子設定はV33を踏襲
    mesh_original = trimesh.load_mesh(input_path)
    center = mesh_original.bounds.mean(axis=0)
    mesh_original.apply_translation(-center)

    total_rotations = random.uniform(0.1, 10.0)
    total_angle_radians = total_rotations * 2 * np.pi
    angle_per_frame = total_angle_radians / FRAME_COUNT

    # 2. カメラの基本姿勢の定義
    camera = pyrender.PerspectiveCamera(yfov=np.pi / 3.0, aspectRatio=1.0)
    T_base = trimesh.transformations.translation_matrix([3.0, -13.0, CAMERA_DISTANCE])
    Rz = trimesh.transformations.rotation_matrix(np.radians(-45), [0, 0, 1])
    Rx = trimesh.transformations.rotation_matrix(np.radians(35.264), [1, 0, 0])
    base_camera_pose = trimesh.transformations.concatenate_matrices(T_base, Rx, Rz)

    # ⚠️ 修正点 2: シーンのアンビエント光をゼロに設定
    scene = pyrender.Scene(bg_color=np.ones(4), ambient_light=np.zeros(3))

    # ⚠️ 修正点 3: 粒子素材をマットにし、Pyrender形式に変換
    material = pyrender.MetallicRoughnessMaterial(
        metallicFactor=0.0,
        roughnessFactor=0.9, # 非常にマットな表面
        baseColorFactor=[0.5, 0.5, 0.5, 1.0] # 灰色
    )
    mesh_pyrender = pyrender.Mesh.from_trimesh(mesh_original, material=material, smooth=False)
    node_mesh = scene.add(mesh_pyrender)

    # ⚠️ 修正点 4: ライトの方向をカメラの真後ろに設定 (Z軸マイナス方向)
    light = pyrender.DirectionalLight(color=np.ones(3), intensity=LIGHT_INTENSITY)
    # カメラはZ軸+方向から見ているため、ライトはZ軸-方向に当たるポーズで配置
    light_pose = trimesh.transformations.translation_matrix([0, 0, 0])
    light_pose = trimesh.transformations.rotation_matrix(np.radians(180), [1, 0, 0]) # ライトの向きを完全に反転

    camera_node = scene.add(camera, pose=base_camera_pose)
    scene.add(light, pose=light_pose) # ライトのポーズを適用

    out = cv2.VideoWriter(output_video_path, fourcc, FPS, RESOLUTION, isColor=False)

    current_rotation_matrix = trimesh.transformations.identity_matrix()
    rotation_axis = np.array([random.uniform(-1, 1), random.uniform(-1, 1), random.uniform(-1, 1)])
    rotation_axis /= np.linalg.norm(rotation_axis)

    NOISE_STD_DEV = random.uniform(0, 3)
    BLUR_KERNEL_SIZE = random.choice([1, 3])

    for frame_index in tqdm(range(FRAME_COUNT), desc=f"Video {video_index+1}"):

        # --- 3D空間での処理 ---
        rotation_matrix_delta = trimesh.transformations.rotation_matrix(angle_per_frame, rotation_axis)
        current_rotation_matrix = trimesh.transformations.concatenate_matrices(current_rotation_matrix, rotation_matrix_delta)
        scene.set_pose(node_mesh, current_rotation_matrix)

        # カメラの揺れ（トラッキング誤差）
        offset_x_px = np.random.normal(0, TRACKING_ERROR_SIGMA)
        offset_y_px = np.random.normal(0, TRACKING_ERROR_SIGMA)
        translation_offset = np.array([offset_x_px / RESOLUTION[0] * 1, offset_y_px / RESOLUTION[1] * 1, 0.0])
        noise_matrix = trimesh.transformations.translation_matrix(translation_offset)
        noisy_camera_pose = trimesh.transformations.concatenate_matrices(noise_matrix, base_camera_pose)
        scene.set_pose(camera_node, noisy_camera_pose)

        # --- レンダリング実行と後処理 ---

        if random.random() < 0.05:
            # 欠損フレーム
            img_gray = np.ones(RESOLUTION, dtype=np.uint8) * 255
        else:
            color, depth = r.render(scene)

            # 1. グレースケール化
            img_gray = cv2.cvtColor(color, cv2.COLOR_RGB2GRAY)

            # 2. ノイズとブラーの適用 (グレースケール画像に直接適用)
            noise = np.random.normal(0, NOISE_STD_DEV, RESOLUTION).astype(np.uint8)
            img_gray = cv2.add(img_gray, noise)

            if BLUR_KERNEL_SIZE > 1:
                img_gray = cv2.GaussianBlur(img_gray, (BLUR_KERNEL_SIZE, BLUR_KERNEL_SIZE), 0)

            # 3. コントラスト強調と二値化 (ノイズがあるため)
            # img_gray = cv2.equalizeHist(img_gray)
            # _, img_binary = cv2.threshold(img_gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

            # # 最終的な出力 (粒子が黒, 背景が白)
            # img_gray = img_binary
            # --- タスク2.1 (動画生成) の書き出しループ内 ---
#
            # 3. コントラスト強調と二値化
            img_gray = cv2.equalizeHist(img_gray)
            _, img_binary = cv2.threshold(img_gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

            # ⚠️ 修正: 先生の指示通り「背景=0(黒), 粒子=255(白)」にする
            # Pyrenderの出力が背景白なら反転させる
            if np.mean(img_binary) > 127:
                img_binary = 255 - img_binary

            img_gray = img_binary
#

        # 動画に書き出し
        out.write(img_gray)

    # 7. 動画書き出しの終了とRendererの解放
    out.release()

    print(f"✅ 動画生成成功: {output_video_path}")

r.delete()
print("\n--- タスク2.1 (ノイズ特化データ生成) 完了 ---")
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
import pandas as pd
import os
import cv2
import numpy as np
from typing import Tuple, List
import torch.nn as nn
import torch.nn.functional as F

# ⚠️ GLOBAL_FEATURE_KEYS は定義されている前提です。

# --- 1. データローダーの実装 (動的正規化とゼロ分散回避の強化) ---
class ParticleDataset(Dataset):
    def __init__(self, video_dir: str, label_path: str):
        self.video_dir = video_dir
        self.labels_df = pd.read_csv(label_path)
        self.video_files = sorted([f for f in os.listdir(video_dir) if '_v3' in f and 'noise_rot' in f])

        self.FEATURE_KEYS = GLOBAL_FEATURE_KEYS

        # 1. ラベルデータ全体からMIN/MAXを動的に計算
        raw_data = self.labels_df[self.FEATURE_KEYS].values

        self.MIN_VALUES = np.min(raw_data, axis=0)
        self.MAX_VALUES = np.max(raw_data, axis=0)

        # 2. ゼロ除算回避の安全策を強化
        self.RANGE_VALUES = self.MAX_VALUES - self.MIN_VALUES

        # 範囲が極端に小さい（例: 1e-8未満）場合も、除算を1.0で行う
        self.RANGE_VALUES[self.RANGE_VALUES < 1e-8] = 1.0


    def __len__(self) -> int:
        return len(self.video_files)

    def __getitem__(self, idx: int) -> Tuple[torch.Tensor, torch.Tensor]:
        video_filename = self.video_files[idx]
        video_path = os.path.join(self.video_dir, video_filename)

        # --- 1. 動画ファイルの読み込みと正規化 --- (省略 - 変更なし)
        cap = cv2.VideoCapture(video_path)
        frames: List[np.ndarray] = []
        while cap.isOpened():
            ret, frame = cap.read()
            if not ret or len(frames) >= 100: break
            frame_gray = frame[:, :, 0]
            frame_normalized = frame_gray.astype(np.float32) / 255.0
            frames.append(frame_normalized)
        cap.release()

        if len(frames) < 100:
             pad_needed = 100 - len(frames)
             padding = np.zeros((pad_needed, 100, 100), dtype=np.float32)
             video_tensor = np.concatenate(frames + [padding], axis=0)
        else:
             video_tensor = np.stack(frames, axis=0)
        video_tensor = torch.from_numpy(video_tensor).unsqueeze(0)


        # --- 2. ラベルの取得と9次元正規化 ---
        parts = video_filename.split('_')
        base_filename_parts = [p for p in parts if p.startswith('particle') or p.startswith('S') or p.startswith('I') or p.startswith('L') or p.startswith('um')]
        base_filename = '_'.join(base_filename_parts)
        label_row = self.labels_df[self.labels_df['File_Name'].str.contains(base_filename + '_Irr.stl', na=False)]

        if label_row.empty:
            raise IndexError(f"ラベルが見つかりません: {base_filename}_Irr.stl")

        raw_features = label_row[GLOBAL_FEATURE_KEYS].iloc[0].values

        # 動的に線形正規化: (X - MIN) / RANGE
        normalized_features = (raw_features - self.MIN_VALUES) / self.RANGE_VALUES

        # 0未満や1超えにならないようにクリップ
        normalized_features = np.clip(normalized_features, 0.0, 1.0)

        labels = torch.tensor(normalized_features, dtype=torch.float32)

        return video_tensor, labels

# --- 2. NNアーキテクチャの定義 (出力サイズを9に変更) ---
class ParticleShapeNet(nn.Module):
    def __init__(self):
        super(ParticleShapeNet, self).__init__()
        self.conv1 = nn.Conv3d(in_channels=1, out_channels=32, kernel_size=3, padding=1)
        self.pool1 = nn.MaxPool3d(kernel_size=2, stride=2)
        self.conv2 = nn.Conv3d(in_channels=32, out_channels=64, kernel_size=3, padding=1)
        self.global_pool = nn.AdaptiveAvgPool3d(1)
        self.fc1 = nn.Linear(in_features=64, out_features=64)

        # 出力サイズを GLOBAL_FEATURE_KEYSの長さ（9）に設定
        self.fc_output = nn.Linear(in_features=64, out_features=len(GLOBAL_FEATURE_KEYS))

    def forward(self, x):
        x = F.relu(self.conv1(x))
        x = self.pool1(x)
        x = F.relu(self.conv2(x))
        x = self.global_pool(x).squeeze(-1).squeeze(-1).squeeze(-1)
        x = F.relu(self.fc1(x))
        x = torch.sigmoid(self.fc_output(x))
        return x
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split
from tqdm import tqdm
import os
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import torch.nn.functional as F

# --- 訓練の実行設定 ---
PROJECT_ROOT = "/content/drive/MyDrive/SHAPE_AI_Project"
VIDEO_DIR = os.path.join(PROJECT_ROOT, '03_Simulation_Videos')
# ⚠️ 修正点 3: ラベルパスを包括的なCSVファイルに修正
LABEL_PATH = os.path.join(PROJECT_ROOT, '02_GroundTruth_Labels', 'comprehensive_particle_labels.csv')

# ハイパーパラメータ
NUM_EPOCHS = 15 # 複雑な学習のためエポック数を増加推奨
BATCH_SIZE = 4
LEARNING_RATE = 0.001
# ⚠️ 修正点 4: モデル保存名をマルチターゲット用に変更
MODEL_SAVE_PATH = os.path.join(PROJECT_ROOT, 'trained_model_multitarget_9dim_final.pth')

# データセットとデータローダーの初期化
dataset = ParticleDataset(VIDEO_DIR, LABEL_PATH)

# データセットの分割 (訓練: 80%, 検証: 20%)
dataset_size = len(dataset)
val_split = 0.2
val_size = int(val_split * dataset_size)
train_size = dataset_size - val_size
train_dataset, _ = random_split(
    dataset,
    [train_size, val_size],
    generator=torch.Generator().manual_seed(42)
)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
print(f"✅ 訓練データセットサイズ: {len(train_dataset)} 個")

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# ⚠️ 注意: ParticleShapeNetの__init__内でParticleDataset.FEATURE_KEYSを参照しているため、
# ParticleDatasetの定義後にParticleShapeNetをインスタンス化する必要があります。
model = ParticleShapeNet().to(device)
# 損失関数は9次元のベクトルの要素ごとのMSEを計算
criterion = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)

print("\n--- モデル訓練を開始します (タスク 3.1 - 9次元マルチターゲット回帰) ---")
model.train()

for epoch in range(NUM_EPOCHS):
    running_loss = 0.0

    for inputs, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{NUM_EPOCHS}"):

        inputs = inputs.to(device)
        labels = labels.to(device) # 9次元ラベル

        optimizer.zero_grad()
        outputs = model(inputs) # 出力も9次元ベクトル

        # 損失の計算 (9次元ベクトル同士のMSE)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * inputs.size(0)

    epoch_loss = running_loss / len(train_loader.dataset)
    print(f"✅ Epoch {epoch+1} 完了. 平均損失 (MSE Loss): {epoch_loss:.6f}")

# --- モデルの保存 ---
torch.save(model.state_dict(), MODEL_SAVE_PATH)
print(f"\n✅ 訓練済みモデルを保存しました: {MODEL_SAVE_PATH}")
```

```python
import torch
import numpy as np
from torch.utils.data import DataLoader, random_split
import torch.nn.functional as F
from tqdm import tqdm
import os
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd

# ------------------- 設定とパス -------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
PROJECT_ROOT = "/content/drive/MyDrive/SHAPE_AI_Project"
MODEL_SAVE_PATH = os.path.join(PROJECT_ROOT, 'trained_model_multitarget_9dim_final.pth')
VIDEO_DIR = os.path.join(PROJECT_ROOT, '03_Simulation_Videos')
LABEL_PATH = os.path.join(PROJECT_ROOT, '02_GroundTruth_Labels', 'comprehensive_particle_labels.csv')
# -------------------------------------------------------------

# データセットの初期化 (ParticleDatasetの定義は実行済みであること)
dataset = ParticleDataset(VIDEO_DIR, LABEL_PATH)
dataset_size = len(dataset)
val_split = 0.2
val_size = int(val_split * dataset_size)
train_size = dataset_size - val_size

_, val_dataset = random_split(
    dataset,
    [train_size, val_size],
    generator=torch.Generator().manual_seed(42)
)

val_loader = DataLoader(val_dataset, batch_size=4, shuffle=False)
print(f"✅ 検証データセットサイズ: {len(val_dataset)} 個")

# --- モデルのロードと検証モードへの設定 ---
model = ParticleShapeNet().to(device)
try:
    model.load_state_dict(torch.load(MODEL_SAVE_PATH))
    print(f"✅ モデルロード成功: {MODEL_SAVE_PATH}")
except FileNotFoundError:
    print(f"❌ エラー: モデルファイルが見つかりません。訓練が完了しているか確認してください。")
    raise
model.eval()

# --- 検証ループの実行 (タスク 3.2) ---
print("\n--- モデル検証を開始します (タスク 3.2: 安定化評価) ---")
all_predictions = []
all_targets = []
total_loss = 0.0

with torch.no_grad():
    for inputs, labels in tqdm(val_loader, desc="Validating"):
        inputs = inputs.to(device)
        labels = labels.to(device)

        outputs = model(inputs)

        loss = F.mse_loss(outputs, labels)
        total_loss += loss.item() * inputs.size(0)

        all_predictions.extend(outputs.cpu().numpy())
        all_targets.extend(labels.cpu().numpy())

# --- 結果の集計と報告 ---
avg_mse = total_loss / len(val_dataset)

predictions_np = np.array(all_predictions)
targets_np = np.array(all_targets)

print("\n--- 検証結果 ---")
print(f"全体平均損失 (MSE): {avg_mse:.6f}")
print(f"全体 RMSE (平方根誤差): {np.sqrt(avg_mse):.6f}")
print("------------------------------")

print("**特性別 詳細評価 (RMSE/R^2)**:")

for i, key in enumerate(GLOBAL_FEATURE_KEYS):
    target_i = targets_np[:, i]
    pred_i = predictions_np[:, i]

    mse_i = mean_squared_error(target_i, pred_i)
    rmse_i = np.sqrt(mse_i)

    # R^2の安定化計算: 分散が極端に小さい場合はR^2を計算しないか、0と見なす
    variance_i = np.var(target_i)

    if variance_i < 1e-7: # 分散が極端に小さい場合 (不安定性の閾値)
        r2_i = 'N/A (分散不安定)'
    else:
        r2_i = r2_score(target_i, pred_i)
        r2_i = f"{r2_i:.4f}"

    # 整形して出力
    print(f"  - {key:<25}: RMSE={rmse_i:.6f}, R^2={r2_i}")

print("\n✅ タスク3.2 (安定化検証) 完了。")
```

```python
import os
import torch
import numpy as np
import pandas as pd
import cv2
import trackpy as tp
import pims
from tqdm import tqdm

# --- 1. モデル定義 (既存のアーキテクチャ) ---
class ParticleShapeNet(torch.nn.Module):
    def __init__(self):
        super(ParticleShapeNet, self).__init__()
        self.conv1 = torch.nn.Conv3d(1, 32, kernel_size=3, padding=1)
        self.pool1 = torch.nn.MaxPool3d(2, 2)
        self.conv2 = torch.nn.Conv3d(32, 64, kernel_size=3, padding=1)
        self.global_pool = torch.nn.AdaptiveAvgPool3d(1)
        self.fc1 = torch.nn.Linear(64, 64)
        self.fc_output = torch.nn.Linear(64, 9)

    def forward(self, x):
        x = torch.nn.functional.relu(self.conv1(x))
        x = self.pool1(x)
        x = torch.nn.functional.relu(self.conv2(x))
        x = self.global_pool(x).view(-1, 64)
        x = torch.nn.functional.relu(self.fc1(x))
        return torch.sigmoid(self.fc_output(x))

# --- 2. 前処理ユーティリティ ---
def extract_roi(frame, x, y, size=100):
    """trackpyの座標(x,y)を中心に100x100でクロップし正規化する"""
    half = size // 2
    y1, y2 = int(y - half), int(y + half)
    x1, x2 = int(x - half), int(x + half)

    # 枠外処理
    if y1 < 0 or x1 < 0 or y2 > frame.shape[0] or x2 > frame.shape[1]:
        return None

    roi = frame[y1:y2, x1:x2]
    # 二値化（既存ロジックを継承）
    _, roi_bin = cv2.threshold(roi, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
    # 背景黒、粒子白に調整（反転が必要な場合は 255 - roi_bin）
    if np.mean(roi_bin) > 127:
        roi_bin = 255 - roi_bin

    return roi_bin.astype(np.float32) / 255.0

# --- 3. メイン解析関数 ---
def run_integrated_analysis():
    # パス設定
    PROJECT_ROOT = "/content/drive/MyDrive/SHAPE_AI_Project"
    MODEL_PATH = os.path.join(PROJECT_ROOT, 'trained_model_multitarget_9dim_final.pth')
    FRAME_DIR = os.path.join(PROJECT_ROOT, "full_video_frames/*.png")
    LABEL_PATH = os.path.join(PROJECT_ROOT, '02_GroundTruth_Labels', 'comprehensive_particle_labels.csv')

    GLOBAL_FEATURE_KEYS = [
        'Wadell_Sphericity', 'Convexity', 'Krumbein_Sphericity',
        'Zingg_S_I_Ratio', 'Zingg_I_L_Ratio',
        'Potticary_Flatness', 'Potticary_Elongation',
        'Tensor_Compactness', 'Tensor_Flakiness'
    ]

    # デバイス準備
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model = ParticleShapeNet().to(device)
    model.load_state_dict(torch.load(MODEL_PATH, map_location=device))
    model.eval()

    # 正規化パラメータの復元
    df_l = pd.read_csv(LABEL_PATH)
    raw_vals = df_l[GLOBAL_FEATURE_KEYS].values
    mins = np.min(raw_vals, axis=0)
    ranges = np.max(raw_vals, axis=0) - mins
    ranges[ranges == 0] = 1.0

    # A. Trackpyによるトラッキング
    print("🚀 ステップ1: Trackpyによる粒子検出とリンクを開始...")
    frames = pims.open(FRAME_DIR)

    # 検出 (diameterは粒子の大きさに合わせて奇数で調整)
    f = tp.batch(frames, diameter=15, minmass=500, invert=False)

    # リンク (search_rangeはフレーム間の最大移動量)
    t = tp.link_df(f, search_range=30, memory=5)

    # 短い軌跡を排除 (最低20フレーム以上存在するもの)
    t_filtered = tp.filter_stubs(t, 20)
    print(f"✅ トラッキング完了。検出粒子数: {t_filtered['particle'].nunique()}")

    # B. 軌跡ごとにROIを抽出してAI判定
    print("🧠 ステップ2: 個別粒子の形状特性を推定中...")
    results = []

    for pid, group in tqdm(t_filtered.groupby('particle')):
        particle_frames = []

        # 軌跡に沿って画像を収集
        for _, row in group.iterrows():
            frame_idx = int(row['frame'])
            roi = extract_roi(frames[frame_idx], row['x'], row['y'])
            if roi is not None:
                particle_frames.append(roi)

        # AI入力用に100フレームにリサイズ/パディング
        if len(particle_frames) < 10: continue

        arr = np.array(particle_frames)
        # 100フレーム分に調整（足りない場合はリピート、多い場合はサンプリング）
        indices = np.linspace(0, len(arr)-1, 100).astype(int)
        input_stack = arr[indices]

        # Tensor変換 (Batch, Channel, Depth, Height, Width)
        inp = torch.FloatTensor(input_stack).unsqueeze(0).unsqueeze(0).to(device)

        with torch.no_grad():
            pred = model(inp).cpu().numpy()[0]

        # 逆正規化して結果を保存
        real_values = (pred * ranges) + mins
        res_entry = {"particle_id": pid, "track_length": len(particle_frames)}
        res_entry.update(dict(zip(GLOBAL_FEATURE_KEYS, real_values)))
        results.append(res_entry)

    # C. 結果保存
    output_df = pd.DataFrame(results)
    output_csv = os.path.join(PROJECT_ROOT, 'trackpy_shape_analysis_results.csv')
    output_df.to_csv(output_csv, index=False)
    print(f"✅ 全工程完了！結果を保存しました: {output_csv}")

if __name__ == "__main__":
    run_integrated_analysis()
```

```python
import cv2
import os

def save_all_frames_to_drive(video_path, output_dir):
    # 1. 動画を読み込む
    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        print(f"❌ 動画ファイルを開けませんでした: {video_path}")
        return

    # 2. 保存先のディレクトリを作成
    os.makedirs(output_dir, exist_ok=True)

    # 総フレーム数を取得して桁数を決める
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    digit = len(str(total_frames))

    print(f"🔄 解析開始: 合計 {total_frames} フレームを画像化します...")

    n = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # ファイル名を生成（例: frame_001.png）
        file_name = f"frame_{str(n).zfill(digit)}.png"
        save_path = os.path.join(output_dir, file_name)

        # 画像を保存
        cv2.imwrite(save_path, frame)

        n += 1

        # 50フレームごとに進捗を表示
        if n % 50 == 0:
            print(f"📈 {n} / {total_frames} 枚保存済み...")

    cap.release()
    print(f"✅ 完了！{n} 枚の画像を以下の場所に保存しました:\n{output_dir}")

# --- 実行設定（あなたの動画パスに合わせています） ---
PROJECT_ROOT = "/content/drive/MyDrive/SHAPE_AI_Project"
input_video = os.path.join(PROJECT_ROOT, "MetcoClad718_d1_15pipe_0_3Lpm_example.mp4")
output_folder = os.path.join(PROJECT_ROOT, "full_video_frames") # わかりやすい名前のフォルダ

save_all_frames_to_drive(input_video, output_folder)
```
